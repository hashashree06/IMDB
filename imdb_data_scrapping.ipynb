```python
# Data cleaning,conversion

import re
import pandas as pd
df=pd.read_csv("merged_data.csv")
# df.info() 
# df.columns

#removing of special characters inside the column
df['MOVIE NAME'] = df['MOVIE NAME'].str.replace(r'^\d+\.\s*', '', regex=True)
#df['VOTING COUNT']=df['VOTING COUNT'].replace('(',"").replace("K","").replace(")","")
#df['VOTING COUNT']=df['VOTING COUNT'].str.replace(r'[\(\)KkMm]', '', regex=True)
df['VOTING COUNT']=df['VOTING COUNT'].str.replace(r'[\(\)]', '', regex=True)
df['DURATION'] = df['DURATION'].str.replace('+', 'm')


#function for converting K to 1000's and M to 1000000
def conversion(x):
    if isinstance(x, str):
        x = x.strip().lower() # Removing the spaces and make the string lowercase
        try:
            if x.endswith('k'):
                return int(float(x[:-1]) * 1_000)
            elif x.endswith('m'):
                return int(float(x[:-1]) * 1_000_000)
            else:
                return int(float(x))
        except:
            print(f'no conversion')
            return None
    return x

#function to convert hours to minutes
def minutes(y):
    hours = minutes = 0
    try:
        h_search = re.search(r'(\d+)\s*h',y)
        m_search = re.search(r'(\d+)\s*m',y)
    
        if h_search:
            hours = int(h_search.group(1)) #group(1) specifically refers to the first capturing group
        if m_search:
            minutes = int(m_search.group(1))
    
        return hours * 60 + minutes

    except:
        return 120
    

# Apply the function
df['DURATION'] = df['DURATION'].apply(minutes)

# Apply the function
df['VOTING COUNT'] = df['VOTING COUNT'].apply(conversion)

print(df)
```
